```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("data/merged.csv")
df
```

```python
def print_na(df):
    # Which columns have missing data
    for _ in df.columns:
        if any(df[_].isna()):
            print(_)

print_na(df)
```

```python
cols = ["max_games_in_a_row","games_won","tiebreaks_won"]
h = "_home"
a = "_away"
for col in cols:
    # when this col is not none, the home and away are both none
    print(df[~df[col].isna()][[col+h,col+a]])
    df[col+h] = df[col+h].fillna(df[col])
    df[col+a] = df[col+a].fillna(df[col])
```

```python
# Drop rank_home and rank_away bc we have rs and seed

df = df.drop(columns = ['rank_home','rank_away']+cols)
```

```python
df.head()
```

```python
df = df[~df["rs"+h].isna() & ~df["rs"+a].isna()]
```

```python
df=df.rename(columns={"win_1":"win_home"})
```

```python
df["avg_set_diff"] = df[['set'+str(x)+'_diff' for x in range(1,6)]].mean(skipna=True, axis=1) # average of game difference across all sets
df["avg_set_games"] = df[['set'+str(x)+'_games' for x in range(1,6)]].mean(skipna=True, axis=1) # average of game difference across all sets
```

Drop rows that have set length NA for 2 for all sex and 3 for men AND any rows where set > 120 minutes

```python
df = df[~df['set2_length'].isna()]
df = df[(~df['set3_length'].isna() & df['men']) | ~df['men']]

sets = df[['set1_length','set2_length','set3_length','set4_length','set5_length']]
df = df[~sets.apply(lambda row: any(row>120), axis=1)]
```

```python
df = df.drop(columns = ['set'+str(x)+'_diff' for x in range(1,6)]+['set'+str(x)+'_length' for x in range(1,6)]+['set'+str(x)+'_games' for x in range(1,6)]+['avg_game_length_set'+str(x) for x in range(1,6)])
```

```python
df['seed'+h] = df['seed'+h].astype(int)
df['seed'+a] = df['seed'+a].astype(int)
```
